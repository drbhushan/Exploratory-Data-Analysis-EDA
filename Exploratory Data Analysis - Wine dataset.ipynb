```python
# Pandas dataframe. corr() is used to find the pairwise correlation of all columns in the dataframe.
plt.figure(figsize=(12,10))
sns.heatmap(data.corr(),annot=True,linewidth=0.5,center=0,cmap='coolwarm')
plt.show()
```

##### In the above graph :
1. Maroon means positive (A positive correlation is a relationship between two variables in which both variables move in the same direction.)
2. Blue means negative (Negative correlation is a relationship between two variables in which one variable increases as the other decreases, and vice versa)


```python
# printing only lower diagional
plt.figure(figsize=(12,10))
mask = np.triu(np.ones_like(data.corr(), dtype=np.bool))
sns.heatmap(data.corr(),annot=True,mask=mask,linewidth=0.5,center=0,cmap='coolwarm')
plt.show()
```

```python
# printing only upper diagional
plt.figure(figsize=(12,10))
mask = np.triu(np.ones_like(data.corr(), dtype=np.bool))
sns.heatmap(data.corr(),annot=True,mask=~mask,linewidth=0.5,center=0,cmap='coolwarm')
plt.show()
```

The following plot presents the distribution of the quality of the wine, 
Avg. wine quality ranging from 5-7

```python
# density plot
sns.distplot(data['quality'], hist=True, kde=True,
             bins='auto', color = 'darkblue',
             hist_kws={'edgecolor':'darkblue'},
             kde_kws={'linewidth': 4})
plt.show()
```

```python
# histogram plot where each hist is visible
# horizontal hist plot. (orientation = 'horizontal')
plt.hist(data.quality,bins=6,alpha=.5,histtype='bar',ec='black',orientation = 'horizontal')
plt.title('Distribution of the Quality')
plt.xlabel('Quality') # x axis label
plt.ylabel('Count') # x axis label
plt.show()
```

```python
# histogram plot where each hist is visible
# vertical hist plot. (orientation = 'vertical')

plt.hist(data.quality,bins=6,alpha=.5,histtype='bar',ec='black',orientation = 'vertical')
plt.title('Distribution of the Quality')
plt.xlabel('Quality') # x axis label
plt.ylabel('Count') # x axis label
plt.show()
```

```python
# other methods of histtypes
plt.hist(data.quality,bins=6,alpha=.5,histtype='step',ec='blue')
plt.title('Distribution of the Quality')
plt.xlabel('Quality') 
plt.ylabel('Count')
plt.show()

plt.hist(data.quality,bins=6,alpha=.5,histtype='stepfilled',ec='black')
plt.title('Distribution of the Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

From the following plot it is clear that, slight increase in pH with the increase in quality.

It is also clear from correlation hear map i.e., **corr(pH,quality) = 0.02**

```python
sns.boxplot(x='quality',y='pH',data=data,palette='BrBG')
plt.title("Quality vs pH")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data['quality'],data['pH'],palette="RdBu_r")
plt.show()
```

From the following plot it is clear that, more the alcohol in the wine better will be the quality.

It is also clear from correlation hear map i.e., **corr(alcohol,quality) = 0.44**

```python
ax = sns.boxplot(x='quality',y='alcohol',data=data,palette='BrBG')
plt.title("Quality vs Alcohol")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data['quality'],data['alcohol'],palette="RdBu_r")
plt.show()
```
